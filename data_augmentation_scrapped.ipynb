```python
import numpy as np
import pandas as pd
import unidecode
import re 
import os
import pickle
import itertools
import random
from nltk.stem.snowball import SpanishStemmer
from nltk.stem.porter import PorterStemmer
import spacy
from es_lemmatizer import lemmatize

nlp = spacy.load('es_core_news_sm')
nlp.add_pipe(lemmatize, after="tagger")
```

```python
def stem_label(label):
    # stemming
    #print("stemming...")
    stemmer = PorterStemmer()
    stemword = lambda s: stemmer.stem(s)
    label = ' '.join([stemword(stemmed_token) for stemmed_token in label.split(' ')])
    #print(label)
    return label
```

```python
def lemmatize(label):  
    #print("lemmatizing...")
    doc = nlp(label)
    lemmas = [token.lemma_ for token in doc]
    label = ' '.join(lemmas)
    #print(label)
    return label
```

```python
def crop_words(label):
    #keep from 1 to 4 first words
    print("cropping...")
    words = label.split(' ')
    nmb_words_to_keep = np.random.randint(1,5)
    words = words[:nmb_words_to_keep]      
    label = ' '.join(words)
    print(label)
    return label
```

```python
def modify_label(label):
    # crop
    print("original label: {}".format(label))
    to_crop = bool(random.getrandbits(1))
    label = crop_words(label) if to_crop else label
    # stemming
    """
    to_stem = bool(random.getrandbits(1))
    label = stem_label(label) if to_stem else label 
    """  
    # lemmas
    to_lemmatize = bool(random.getrandbits(1))
    label = lemmatize(label) if to_lemmatize else label 
    label = unidecode.unidecode(label)
    print("final label: {}".format(label))
    return label
```

```python
def augmentation_labels(original_series):
    modified_series = original_series.copy()
    modified_series = modified_series.apply(lambda x: modify_label(x))
    series = pd.concat([original_series,modified_series], ignore_index=True, sort=True)  
    series = series.drop_duplicates().reset_index(drop=True)
    print("Total labels in augmentation {}".format(series.count()))
    return series   
```

```python
df = pd.read_csv("../../../data/scrapped/products_spain/alcampo/bakery/bakery/fresh_bakery.csv", header=0)
df.drop(columns=['web-scraper-order','web-scraper-start-url'], inplace=True)
print(df.info())  
if df.count()['label'] < 1000:
    new_labels = augmentation_labels(df['label'])
    df = pd.DataFrame(new_labels) 
df.insert(0,'category_uid', '14')
print(df.info())    
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
def remove_stopwords(label):
    label = ' '.join([word if word not in stopwords.words('spanish') else '' for word in label.split(' ')])
    return label
```
